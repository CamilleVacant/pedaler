### Filtrage des données

```python
df = df[(df['Station en fonctionnement'] == 'OUI') & 
   (df['Nom communes équipées'] == 'Paris') &
   (df['Actualisation de la donnée'].str.startswith('2024'))
   ]
df.shape
```
